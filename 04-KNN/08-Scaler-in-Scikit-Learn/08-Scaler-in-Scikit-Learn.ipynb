# Scikit-learn 中的 Scaler

```python
import numpy as np
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
X = iris.data
y = iris.target
```

```python
X[:10, :]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=666)
```

## scikit-learn 中的 StandardScaler

```python
from sklearn.preprocessing import StandardScaler
```

```python
standradScaler = StandardScaler()
```

```python
standradScaler.fit(X_train)
```

```python
standradScaler.mean_
```

```python
standradScaler.scale_
```

```python
standradScaler.transform(X_train)
```

```python
X_train
```

```python
X_train = standradScaler.transform(X_train)
```

```python
X_train
```

```python
X_test_standard = standradScaler.transform(X_test)
```

```python
X_test_standard
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_clf = KNeighborsClassifier(n_neighbors=3)
```

```python
knn_clf.fit(X_train, y_train)
```

```python
knn_clf.score(X_test_standard, y_test)
```

```python
knn_clf.score(X_test, y_test)
```
