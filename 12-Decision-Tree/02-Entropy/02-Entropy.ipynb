# 信息熵

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def entropy(p):
    return -p * np.log(p) - (1 - p) * np.log(1 - p)
```

```python
x = np.linspace(0.01, 0.99, 200)
```

```python
plt.plot(x, entropy(x))
```

# 基尼系数

```python
def gini(p):
    return 1 - p**2 - (1 - p)**2
```

```python
plt.plot(x, gini(x))
```
